```python
#libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
#read Dataset
df=pd.read_csv('E:\\python for data science\\practical machine learning\\archive\\spam detection dataset\\spam_mail_data.csv')
```

```python
#check if the data was uploaded
df.head(10)
```

```python
df.groupby('Category').describe()
```

```python
df['spam']=df['Category'].apply(lambda x:1 if x=='spam' else 0)
```

```python
df.head()
```

```python
#convert Message column to a numericat values using
# CountVectorizer Technique
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df.Message, df.spam,  test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vec=CountVectorizer()
```

```python
x_train_count=vec.fit_transform(x_train.values)
x_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_count, y_train)
```

```python
#calculate the score of the model , here we'll convert the x_test into a count
x_test_count = vec.transform(x_test)
model.score(x_test_count, y_test)
```

```python
#test the model 
sample =[ 'Ahhh. Work. I vaguely remember that! What does it feel like? Lol']
# here we'll convert that e-mail into a countVectorizer
sample_count = vec.transform(sample)
model.predict(sample_count)
```

```python
#this sample is a spam , lets see that on the model
spam_test = ['Congrats! 1 year special cinema pass for 2 is yours. call 09061209465 now! C Suprman V, Matrix3, StarWars3, etc all 4 FREE! bx420-ip4-5we. 150pm. Dont miss out!']
spam_test_count = vec.transform(spam_test)
model.predict(spam_test_count)
```

```python
#that's all , now let's save the model
import pickle
with open ('Spam Detectoon','wb') as f:
    pickle.dump(model, f)
```
